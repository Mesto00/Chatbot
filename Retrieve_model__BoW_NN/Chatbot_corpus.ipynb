On ne garde que les films américains, français et italien pour notre cas d'usage car plus populaires

```python
countries = ["USA", "France", "Italy"]
data = data[data.country.isin(countries)]
```

```python
data = data.drop(['usa_gross_income', 'worlwide_gross_income',"metascore", "title" ], axis=1)
```

```python
data = data.rename(columns={"imdb_title_id":"imdb title id", "original_title": "title" , "date_published":"date",
                            "production_company":"producer","avg_vote":"average of votes", "votes":"number of votes",
                            "reviews_from_users":"number of reviews from users","reviews_from_critics":"number of reviews from critics"})
```
